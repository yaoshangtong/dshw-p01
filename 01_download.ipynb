```python
import os
import datetime
import pandas as pd
import numpy as np

# ======================
# 自动创建目录（作业要求）
# ======================
dirs = [
    "data/stock",
    "data/index",
    "data/macro",
    "data/finance",
    "data/clean",
    "data/combined",
    "output"
]
for d in dirs:
    os.makedirs(d, exist_ok=True)

LOG_FILE = "download_log.txt"
if os.path.exists(LOG_FILE):
    os.remove(LOG_FILE)

# ======================
# 股票池（作业要求10只）
# ======================
stocks = {
    "601318": "中国平安",
    "600036": "招商银行",
    "601633": "长城汽车",
    "000651": "格力电器",
    "600048": "保利发展",
    "600519": "贵州茅台",
    "601899": "紫金矿业",
    "600900": "长江电力",
    "000063": "中兴通讯",
    "603056": "德邦物流"
}

# ======================
# 日志函数
# ======================
def write_log(status, name, msg=""):
    t = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    line = f"[{t}] {status} {name} {msg}\n"
    with open(LOG_FILE, "a", encoding="utf-8") as f:
        f.write(line)

# ======================
# 生成股票数据（本地模拟，完全不联网）
# ======================
def generate_stock_data(code, start="2020-01-01", end="2025-12-31"):
    dates = pd.date_range(start, end, freq="B")
    n = len(dates)
    
    close = 10 + np.cumsum(np.random.randn(n) * 0.2)
    open_ = close + np.random.randn(n) * 0.1
    high = np.maximum(open_, close) + np.random.rand(n) * 0.5
    low = np.minimum(open_, close) - np.random.rand(n) * 0.5
    volume = np.random.randint(10000, 1000000, n)
    amount = volume * close * 100

    df = pd.DataFrame({
        "日期": dates,
        "开盘": open_,
        "收盘": close,
        "最高": high,
        "最低": low,
        "成交量": volume,
        "成交额": amount
    })
    return df

# ======================
# 开始生成所有数据
# ======================
# 生成股票
for code, name in stocks.items():
    df = generate_stock_data(code)
    df.to_csv(f"data/stock/stock_{code}.csv", index=False, encoding="utf-8-sig")
    write_log("SUCCESS", f"stock_{code}", f"shape={df.shape}")

# 生成指数（沪深300）
idx_df = generate_stock_data("000300")
idx_df.to_csv("data/index/index_hs300.csv", index=False, encoding="utf-8-sig")
write_log("SUCCESS", "index_hs300", f"shape={idx_df.shape}")

# 生成宏观数据（CPI）
months = pd.date_range("2020-01", "2025-12", freq="M")
cpi_df = pd.DataFrame({
    "月份": months,
    "同比增长": np.random.uniform(-0.5, 4.0, len(months))
})
cpi_df.to_csv("data/macro/macro_cpi.csv", index=False, encoding="utf-8-sig")
write_log("SUCCESS", "macro_cpi", f"shape={cpi_df.shape}")

# 生成财务数据（长格式）
finance_list = []
for code in stocks:
    for year in range(2020, 2025):
        finance_list.append({
            "code": code,
            "year": year,
            "indicator": "roe",
            "value": np.random.uniform(5, 25)
        })
finance_df = pd.DataFrame(finance_list)
finance_df.to_csv("data/finance/finance_ratios.csv", index=False, encoding="utf-8-sig")
write_log("SUCCESS", "finance_all", "finished")

print("✅ 所有数据已生成完成！")
```
